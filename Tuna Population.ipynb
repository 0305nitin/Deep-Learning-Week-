```python
import numpy as np

# Parameters
r_annual = 0.1  # Annual intrinsic rate of increase
K = 10000  # Carrying capacity
days = np.arange(366)  # 0 to 365 days
initial_populations = np.arange(100, 1100, 100)  # Initial populations from 100 to 1000 in increments of 100

# Convert annual growth rate to daily growth rate
r_daily = (1 + r_annual) ** (1/365) - 1

# Logistic growth function
def logistic_growth(P0, r, K, t):
    return K / (1 + ((K - P0) / P0) * np.exp(-r * t))

# Create a 2D array to hold the results
population_growth_daily = np.zeros((len(initial_populations), len(days)))

# Populate the array
for i, P0 in enumerate(initial_populations):
    for j, day in enumerate(days):
        population_growth_daily[i, j] = logistic_growth(P0, r_daily, K, day)

print("Daily Population Growth Over 365 Days:")
print(population_growth_daily)
```

```python
import numpy as np

# Parameters
rows, columns = 25, 25  # Dimensions of the 2D array
K = 1000000  # Carrying capacity
r_annual = 0.1  # Annual intrinsic rate of increase
days = 30  # Number of days to simulate

# Convert annual growth rate to daily growth rate
r_daily = (1 + r_annual) ** (1/365) - 1

# Initialize a 25x25 array with random initial populations between 50 and 1000
np.random.seed(42)  # For reproducible results
initial_populations = np.random.randint(50, 1001, size=(rows, columns))

# Logistic growth function
def logistic_growth(P0, r, K, t):
    return K / (1 + ((K - P0) / P0) * np.exp(-r * t))

# Simulate growth over 30 days
population_growth_daily = np.zeros((rows, columns, days+1))  # +1 to include the initial day
population_growth_daily[:, :, 0] = initial_populations  # Set initial populations

for day in range(1, days+1):
    population_growth_daily[:, :, day] = logistic_growth(initial_populations, r_daily, K, day)

# Example: Display the population on the last day of simulation for the entire array
print("Population on Day 30:")
print(population_growth_daily[:, :, -1])
```

```python
# Display function to show populations for selected days
def display_population_for_selected_days(population_growth_daily, days_to_show):
    for day in days_to_show:
        print(f"Day {day}: Population Growth")
        for row in range(population_growth_daily.shape[0]):
            for col in range(population_growth_daily.shape[1]):
                # Print with spaces for readability, rounding for simplicity
                print(f"{population_growth_daily[row, col, day]:.0f}".rjust(5), end=" ")
            print()  # Newline for next row
        print("\n" + "="*100 + "\n")  # Separator for next day's display

# Define which days to display
days_to_show = [0, 5, 10, 15, 20, 25, 30]

# Call the display function
display_population_for_selected_days(population_growth_daily, days_to_show)
```

```python
# Parameters
rows, columns = 25, 25  # Dimensions of the 2D array
K = 1000000  # Carrying capacity
r_annual = 0.1  # Annual intrinsic rate of increase
months = 24  # Simulate growth over 24 months (2 years)

# Convert annual growth rate to monthly growth rate
r_monthly = (1 + r_annual) ** (1/12) - 1

# Initialize a 25x25 array with random initial populations between 50 and 1000
np.random.seed(42)  # For reproducible results
initial_populations = np.random.randint(1000, 255000, size=(rows, columns))

# Logistic growth function
def logistic_growth(P0, r, K, t):
    return K / (1 + ((K - P0) / P0) * np.exp(-r * t))

# Simulate growth over 24 months
population_growth_monthly = np.zeros((rows, columns, months+1))  # +1 to include the initial month
population_growth_monthly[:, :, 0] = initial_populations  # Set initial populations

for month in range(1, months+1):
    population_growth_monthly[:, :, month] = logistic_growth(initial_populations, r_monthly, K, month)

# Function to display populations for selected months
def display_population_for_selected_months(population_growth_monthly, months_to_show):
    for month in months_to_show:
        print(f"Month {month}: Population Growth")
        for row in range(population_growth_monthly.shape[0]):
            for col in range(population_growth_monthly.shape[1]):
                # Print with spaces for readability, rounding for simplicity
                print(f"{population_growth_monthly[row, col, month]:.0f}".rjust(6), end=" ")
            print()  # Newline for next row
        print("\n" + "="*120 + "\n")  # Separator for next month's display

# Define which months to display
months_to_show = [0, 3, 6, 9, 12, 15, 18, 21, 24]

# Call the display function
display_population_for_selected_months(population_growth_monthly, months_to_show)
```
